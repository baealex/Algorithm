```python
# 행렬의 곱셈: 첫번째 행렬의 열과 두번째 행렬의 행이 같아야함.
# 결과: 첫번째 행렬의 행과 두번째 행렬의 열로 규성됨
v = np.array([[1, 2, 3], [4, 5, 6]])  # 2행 3열
w = np.array([[1, 1], [2, 2], [3, 3]]) # 3행 2열
print('v:', v.shape)
print('w:', w.shape)
print(v)
print(w)
print(v.dot(w))  # 2행 2열, 1 * 1 + 2 * 2 + 3 * 3 --> 1 + 4 + 9 
```

```python
# ERROR
v = np.array([[1, 2]])  # 1행 2열
w = np.array([[1, 1], [2, 2], [3, 3]]) # 3행 2열
print('v:', v.shape)
print('w:', w.shape)
print(v)
print(w)
print(v.dot(w))  # 1행 2열, 1 * 1 + 2 * 2 + 3 * 3 --> 1 + 4 + 9 
```

```python
def my_func(D):
    m = np.mean(D)  # 평균
    s = np.std(D)   # 표준 편차
    return m, s
```

```python
data=np.random.randn(100)
data_mean, data_std = my_func(data)
print('mean:{0:3.2f}, std:{1:3.2f}'.format(data_mean, data_std))
```

```python
output = my_func(data)
print(output)
print(type(output))
print('mean:{0:3.2f}, std:{1:3.2f}'.format(output[0], output[1]))
```
